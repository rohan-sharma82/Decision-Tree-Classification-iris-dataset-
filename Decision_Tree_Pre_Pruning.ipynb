```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score,classification_report
```

```python
#loading dataset
iris=load_iris()
iris
```

```python
#target values we have
iris.target
```

```python
#another method of getting iris dataset
df=sns.load_dataset('iris')
```

```python
#independent and dependent features
X=df.iloc[:,:-1]
Y=iris.target
```

```python
X,Y
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_trian,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=39)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn import tree
```

```python
##preprocessing
parameter={
    'criterion':['gini','entropy'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,6],
    'max_features':['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treemodel=DecisionTreeClassifier()
cv=GridSearchCV(treemodel,param_grid=parameter,cv=5)
```

```python
cv.fit(x_trian,y_train)
```

```python
cv.best_params_
```

```python
y_predict=cv.predict(x_test)
```

```python
score=accuracy_score(y_predict,y_test)
score
```

```python
print(classification_report(y_predict,y_test))
```
