```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score,classification_report
```

```python
#loading dataset
iris=load_iris()
iris
```

```python
#target values we have
iris.target
```

```python
#another method of getting iris dataset
df=sns.load_dataset('iris')
```

```python
#independent and dependent features
X=df.iloc[:,:-1]
Y=iris.target
```

```python
X,Y
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_trian,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=39)
```

```python
#by default it is a post pruning techinque
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(max_depth=3) #max_depth helps to avoid overfitting in the model
dt.fit(x_trian,y_train)
```

```python
#ploting the trees for better understanding
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(dt,filled=True)
```

```python
#prediction
y_pred=dt.predict(x_test)
```

```python
y_pred
```

```python
score=accuracy_score(y_test,y_pred)
score
```

```python
print(classification_report(y_test,y_pred))
```
